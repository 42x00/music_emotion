```python
import pandas as pd

df = pd.read_csv('../../data/label/MeanCategoryRatingsUSA.csv')
df.head()
```

```python
from tqdm import tqdm
import soundfile as sf

sdf = pd.DataFrame({'filename': df['filename']})
sdf['y_length'], sdf['sr'] = 0, 0
sdf.head()
```

```python
for i, row in tqdm(sdf.iterrows(), total=sdf.shape[0]):
    y, sr = sf.read(f'../../data/sample/{row.filename}')
    sdf.at[i, 'y_length'] = len(y)
    sdf.at[i, 'sr'] = sr
```

```python
sdf.head()
```

```python
sdf['length'] = sdf['y_length'] / sdf['sr']
sdf.head()
```

```python
sdf['length'] = sdf['length'].apply(lambda x: round(x))
sdf.head()
```

```python
ssdf = sdf[sdf['length'] == 5]
ssdf.describe()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(ssdf, test_size=0.2)
```
